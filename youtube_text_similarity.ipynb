```
import tensorflow as tf
import tensorflow_hub as hub
import matplotlib.pyplot as plt
import numpy as np
import io
import os
import pandas as pd
import re
import seaborn as sns
```

```
module_url = "https://tfhub.dev/google/universal-sentence-encoder/2" #@param ["https://tfhub.dev/google/universal-sentence-encoder/2", "https://tfhub.dev/google/universal-sentence-encoder-large/3"]
```

```
# Import the Universal Sentence Encoder's TF Hub module
embed = hub.Module(module_url)
```

```
import math

def notnan(x):
  if x == str(x):
    return True
  if math.isnan(x):
    return False
  
  return True

df = pd.read_csv(io.StringIO(uploaded['kma_youtube.csv'].decode('utf-8')))
df = df[df['video_description'] != '']
# df = df[df.captions_text.apply(notnan)]

df.head(10)
```

```
def join_text(row):
  text  = row['video_title']
  text += ' '
  text += str(row['video_description'])
  text += ' '
  if row.get('tags'):
    #print(row.get('tags'))
    text += str(row['tags'])
    text += ' '
  if row.get('captions_text'):
    text += str(row['captions_text'])
  
  t = re.sub(r'\W+\s', ' ', text)
  #print(t)
  return t.replace('"', '')

tvd = {}

def text_vid_hasher(row):
  vid = row['video_url']
  text = join_text(row)
  #print(text)
  #return {text: vid}
  tvd[text] = vid

df.apply(text_vid_hasher, axis=1)
tvd
```

```
messages = list(tvd.keys())

with tf.Session(config=tf.ConfigProto(
      allow_soft_placement=True, log_device_placement=True)) as session:
  session.run([tf.global_variables_initializer(), tf.tables_initializer()])
  message_embeddings = session.run(embed(messages))

  for i, message_embedding in enumerate(np.array(message_embeddings).tolist()):
    if i > 4:
      break
    print("Message: {}".format(messages[i]))
    print("Embedding size: {}".format(len(message_embedding)))
    message_embedding_snippet = ", ".join(
        (str(x) for x in message_embedding[:3]))
    print("Embedding: [{}, ...]\n".format(message_embedding_snippet))
```
